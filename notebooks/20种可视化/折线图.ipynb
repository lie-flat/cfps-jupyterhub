## 可视化

```python
# 数据处理
condi = f'{var}>0'
```

```python
x = [year1,year2,year3]
title = info1['key']
data1 = sql(f'select {var} from {table}_{year1} where {condi}')
data2 = sql(f'select {var} from {table}_{year2} where {condi}')
data3 = sql(f'select {var} from {table}_{year3} where {condi}')
y_data = [int(data1.mean().values[0]),int(data2.mean().values[0]),int(data3.mean().values[0])]
lis = zip(x,y_data)
lis = sorted(lis)
x_data = []
y_data = []
for i in range(len(lis)):
    x_data.append(str(lis[i][0]))
    y_data.append(lis[i][1])
```

```python
(
    Line()
    .set_global_opts(
        tooltip_opts=opts.TooltipOpts(is_show=False),
        xaxis_opts=opts.AxisOpts(type_="category"),
        yaxis_opts=opts.AxisOpts(
            type_="value",
            axistick_opts=opts.AxisTickOpts(is_show=True),
            splitline_opts=opts.SplitLineOpts(is_show=True),
        ),
        title_opts=opts.TitleOpts(f"{info1['key']}(均值)")
    )
    .add_xaxis(xaxis_data=x_data)
    .add_yaxis(
        series_name="",
        y_axis=y_data,
        symbol="emptyCircle",
        is_symbol_show=True,
        label_opts=opts.LabelOpts(is_show=False),
    )
    .render_notebook()
)
```
