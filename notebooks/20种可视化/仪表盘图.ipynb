### 请选择您关心的变量：

```python
dtal = '普通初中' 
```

```python
# 计算出占比
title = info['key']+'——'+dtal+'占比'
def calcu_per(counts,dtal):
    perct = 0
    for i,n in info['range'].items():
        if n == dtal:
            perct = (int(counts[int(i)])/int(counts.sum()))*100
    return ['%.2f'%perct]
    # return perct

per = calcu_per(counts,dtal)
```
