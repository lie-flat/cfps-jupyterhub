## 从上面所列变量中，选择您所关心的

```python
var = 'wn1'
loca = 'provcd'
```

```python
info = cfps[year][table].schema[var]
lc = cfps[year][table].schema[loca]
temp = {}
temp1 = {}
# 对info进行一个剔除筛选：
for i,n in info['range'].items():
    if int(i)>0:
       temp[i] = n
info['range'] = temp

for i,n in lc['range'].items():
    if int(i)>0:
        temp1[i]=n
lc['range'] = temp1
info,lc
```

```python
condi = f'{var}>0 and {loca}>0'
```

```python
df = sql(f'select {loca},{var} from {table}_{year} where {condi}')
# data.value_counts()
```

```python
df1 = df.groupby('provcd')
```

```python
dic = {}
big_dic = {}
for i in range(len(list(df1))):
    for j,n in list(df1)[i][1]['wn1'].value_counts().items():
        dic[j] = n
    big_dic[list(df1)[i][0]] = dic
    dic = {}
```

```python
# 确定y轴坐标有几类：
y = []
for i,n in info['range'].items():
    y.append(n)
y
```

```python
df1 = df.groupby('provcd')
```

```python
dic = {}
big_dic = {}
for i in range(len(list(df1))):
    for j,n in list(df1)[i][1]['wn1'].value_counts().items():
        j = list(info['range'].keys()).index(str(j)) 
        dic[j] = n
    i = list(lc['range'].keys()).index(str(list(df1)[i][0]))
    big_dic[i] = dic
    dic = {}
```

```python
# 再得到三维坐标
data = []
for i,n in big_dic.items():
    for j,m in n.items():
        data.append([i,j,m])
```

```python
x_data = []
title = info['key']
for i,n in lc['range'].items():
    x_data.append(n[:-1])
```

```python
(
    HeatMap()
    .add_xaxis(x_data)
    .add_yaxis("人数",y, data)
    .set_global_opts(
        title_opts=opts.TitleOpts(title=title),
        visualmap_opts=opts.VisualMapOpts(),
        xaxis_opts=opts.AxisOpts(name="省", axislabel_opts={"rotate": -45})
        
    )
    # .set_series_opts(xaxis_opts = opts.LabelOpts(rotate = 70))
    .render_notebook()
)
```
