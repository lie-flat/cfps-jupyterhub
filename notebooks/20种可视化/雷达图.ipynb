添加保留条件，去除某些值（如去除-8等不适用的值）

必须使用合法的SQL表达式，使用`{var}`作为占位符，例如`0<={var}<1000`

如果多个变量的条件一样，可以使用乘法来复制列表元素

```python
conds = ['{var}>=0'] * 6
# conds = ['{var}>0'] * 6
```

如果该变量是数值型，需要插入分点将给定的数值分成若干组，每组记录落在这个区间内的人数

```python
is_numbers = [(schema[var]['type'] in ['int32', 'float64']) for var in variables]
print('变量是否为数值型：', is_numbers)
for i, is_number in enumerate(is_numbers):
    if(is_number):
        print('{}变量的取值范围：'.format(variables[i]), schema[variables[i]]['range'])
```

## 可视化

可定义标题以及添加每个变量的描述

```python
title = '信息了解途径'

names = ['电视', '互联网', '报纸', '广播', '短信', '他人转告']

# title = '孩子看电视情况'

# names = [schema[var]['key'] for var in variables]
```

```python
from math import isnan

descriptions = [schema[var]['key'] for var in variables] # 变量名
pre_data = [sql(f'SELECT {var} FROM {table}_{year} WHERE ' + cond.format(var=var))[var].tolist() for var, cond in zip(variables, conds)]
if(is_number):
    # 对于数值型变量，计算落在该区间的均值
    
    # 过滤nan
    pre_data = [list(filter(lambda x: not isnan(x), item)) for item in pre_data]
    
    # 计算均值
    data = [[sum(item)/len(item) for item in pre_data]]
else:
    # 将分类变量的所有选项存于字典内
    type_dicts = [schema[var]['range'] for var in variables]
    data = [
        [sum([int(y == int(key)) for y in pre_data_item if y > 0]) for key in type_dict if eval(cond.format(var=key))] 
        for pre_data_item, type_dict, cond in zip(pre_data, type_dicts, conds)
    ]
    
#保留3位有效数字
data = [[round(value, 2) for value in data_item] for data_item in data]
```

```python
from itertools import chain

if(is_number):
    radar_schema = [{'name': name, 'max': schema[var]['minmax'][1]} for name, var in zip(names, variables)]
else:
    radar_schema = [{'name': name, 'max': max(chain(data))} for name in names]

radar = (
    Radar()
    .add_schema(radar_schema)
    .add(title, data)
    .set_global_opts(title_opts=opts.TitleOpts(title=title))
)
radar.render_notebook()
```
