# 箱线图

箱线图又称为盒须图，是一种用作显示一组数据分散情况资料的统计图。因形状如箱子而得名。其基本原理是显示一组数据的中位数，1/4分位数，3/4分位数，并计算上下边缘和超出数据的部分。

选择一个变量

```python
variables = ['dress', 'food', 'daily', 'trco']
# variables = ['wf601m', 'wf602m', 'wf603m', 'wf604m', 'wf605m', 'wf606m']
# var = 'fincome1'
```

添加保留条件，去除某些值（如去除-8等不适用的值）

必须使用合法的SQL表达式，使用`{var}`作为占位符，例如`0<={var}<1000`

如果多个变量的条件一样，可以使用乘法来复制列表元素

```python
# conds = ['{var}>=4'] * 6
conds = ['{var}>0'] * 4
```

如果该变量是数值型，需要插入分点将给定的数值分成若干组，每组记录落在这个区间内的人数

```python
is_numbers = [(schema[var]['type'] in ['int32', 'float64']) for var in variables]
print('变量是否为数值型：', is_numbers)
for i, is_number in enumerate(is_numbers):
    if(is_number):
        print('{}变量的取值范围：'.format(variables[i]), schema[variables[i]]['range'])
```

```python
#分点数，按顺序插入（数值型不能为空，若不是数值型可以忽略这一项）
# seps = [1000, 5000, 10000, 50000, 100000, 300000, 500000]
```

## 可视化

可定义标题以及添加每个变量的描述

```python
# title = '信息了解途径'

names = ['衣物支出', '食物支出', '日用品支出', '交通支出']
```

```python
from math import isnan

descriptions = [schema[var]['key'] for var in variables] # 变量名
data = [sql(f'SELECT {var} FROM {table}_{year} WHERE ' + cond.format(var=var))[var].tolist() for var, cond in zip(variables, conds)]
    
#保留3位有效数字
data = [[round(value, 2) for value in data_item] for data_item in data]
for i, item in enumerate(data):
    item.sort()
    data[i] = item
```

```python
boxplot = (
    Boxplot()
    .add_xaxis(names)
)
boxplot.add_yaxis(table, boxplot.prepare_data(data))
boxplot.render_notebook()
```
