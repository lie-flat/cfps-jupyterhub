```python
var = 'wm601code'
```

## 需要保留哪些值 (默认保留大于0的值)

```python
print('\n'.join(f"{k}: {v}" for k,v in schema[var]['range'].items()))
```

```python
cond = f'{var}>0' # 请输入 SQL 条件语句
```

## 可视化

```python
description = schema[var]['key'] # 变量的描述
data = sql(f"SELECT {var} FROM {table}_{year} WHERE {cond};")
counts = data.value_counts().to_dict() # 统计各个值的数目
dic = {schema[var]['range'][str(int(index))]:value for (index,), value in counts.items()} # 生成 文字->数值 映射
```

```python
(
    WordCloud()
    .add(series_name=description, data_pair=dic.items())
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title=description, title_textstyle_opts=opts.TextStyleOpts(font_size=23)
        ),
        tooltip_opts=opts.TooltipOpts(is_show=True),
    )
    .render_notebook()
)
```
