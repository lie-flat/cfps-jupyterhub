选择一个变量

```python
var = 'qa5code'
# var = 'fincome1'
```

添加保留条件，去除某些值（如去除-8等不适用的值）

必须使用合法的SQL表达式，使用`{var}`作为占位符，例如`0<={var}<1000`

```python
cond = f'{var}>=0'
```

如果该变量是数值型，需要插入分点将给定的数值分成若干组，每组记录落在这个区间内的人数

```python
is_number = (schema[var]['type'] in ['int32', 'float64'])
print('变量是否为数值型：', is_number)
if(is_number):
    print('变量的取值范围：', schema[var]['range'])
```

```python
#分点数，按顺序插入（数值型不能为空，若不是数值型可以忽略这一项）
seps = [1000, 5000, 10000, 50000, 100000, 300000, 500000]
```

## 可视化

可定义标题

```python
title = None
```

```python
description = schema[var]['key'] # 变量名
pre_data = sql(f'SELECT {var} FROM {table}_{year} WHERE {cond}')[var].tolist()
if(is_number):
    seps.sort()
    seps = [0] + seps + [float('inf')]
    data = [('[{}, {})'.format(a, b), sum([int(a <= x < b) for x in pre_data]) ) for a, b in zip(seps[:-1], seps[1:])]
else:
    # 将分类变量的所有选项存于字典内
    type_dict = schema[var]['range']
    data = [(type_dict[key], sum([int(y == int(key)) for y in pre_data if y >= 0]) ) for key in type_dict]
    # 去除人数为0的民族
    data = [(x, y) for x, y in data if y>0]
    # 对民族人数降序排列
    data.sort(key=lambda x:x[1], reverse=True)
```

```python
pie = (
    Pie(init_opts=opts.InitOpts(theme=ThemeType.LIGHT))
    .add(series_name=description, data_pair=data)
    .set_global_opts(
        title_opts=opts.TitleOpts(title=title),
        legend_opts=opts.LegendOpts('scroll')
    )
)
pie.render_notebook()
```
