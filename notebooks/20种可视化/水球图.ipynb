### 您可以关心的变量是

```python
dtal = '普通初中' 
```

```python
# 计算出占比
def calcu_per(counts,dtal):
    perct = 0
    for i,n in info['range'].items():
        if n == dtal:
            perct = int(counts[int(i)])/int(counts.sum())
    return ['%.2f'%perct]
per = calcu_per(counts,dtal)
```

```python
title = info['key']+'——'+dtal+'占比'
l1 = (
    Liquid()
    .add("lq", per, center=["30%", "50%"])
    .set_global_opts(title_opts=opts.TitleOpts(title=title))
    # .render_notebook()
)
grid = Grid().add(l1, grid_opts=opts.GridOpts())
grid.render_notebook()
```
