## 请选择两个变量,变量 a 将作为自变量，变量 b 将作为因变量

```python
import re
print("可以选择的变量有:")
schema = cfps[year][table].schema
print(
    '\n'.join(
        [f'{var}: {value["key"]}' for var, value in schema.items() if re.match(r"^(float|int)", value["type"])]
    ) or '错误：没有可以选择的变量，请尝试更改年份和表格'
)
```

```python
var_a = 'ft1'
var_b = 'expense'
```

```python
print(f"您选择了 `{schema[var_a]['key']}` 和 `{schema[var_b]['key']}`")
```

## 需要保留哪些值 （默认保留大于等于0的值）

```python
print(f'{var_a} 的取值范围：{schema[var_a]["minmax"]}')
print(f'{var_b} 的取值范围：{schema[var_b]["minmax"]}')
```

```python
cond_a = f'{var_a} >= 0' # 条件必须是合法的 SQL 条件表达式
cond_b = f'{var_b} >= 0'
```

# 可视化

```python
data = sql(f"SELECT {var_a},{var_b} FROM {table}_{year} WHERE {cond_a} AND {cond_b}")
(
    EffectScatter()
    .add_xaxis(
        xaxis_data=data[var_a]
    )
    .add_yaxis(
        series_name=schema[var_b]["key"],
        y_axis=data[var_b],
        symbol_size=8,
        label_opts=opts.LabelOpts(is_show=False),
    )
    .set_series_opts()
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title=f"{schema[var_b]['key']}~{schema[var_a]['key']}",
            pos_left="center",
            pos_top="top"
        ),
        xaxis_opts=opts.AxisOpts(
            type_="value", splitline_opts=opts.SplitLineOpts(is_show=True)
        ),
        yaxis_opts=opts.AxisOpts(
            type_="value",
            axistick_opts=opts.AxisTickOpts(is_show=True),
            splitline_opts=opts.SplitLineOpts(is_show=True),
        ),
        tooltip_opts=opts.TooltipOpts(is_show=True),
        legend_opts=opts.LegendOpts(pos_top="bottom")
    )
    .render_notebook()
)
```

## 您也可以选择使用对数坐标轴

```python
(
    EffectScatter()
    .add_xaxis(
        xaxis_data=data[var_a]
    )
    .add_yaxis(
        series_name=schema[var_b]["key"],
        y_axis=data[var_b],
        symbol_size=8,
        label_opts=opts.LabelOpts(is_show=False),
    )
    .set_series_opts()
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title=f"{schema[var_b]['key']}~{schema[var_a]['key']}",
            pos_left="center",
            pos_top="top"
        ),
        xaxis_opts=opts.AxisOpts(
            type_="log",min_=1, splitline_opts=opts.SplitLineOpts(is_show=True)
        ),
        yaxis_opts=opts.AxisOpts(
            type_="log", min_=1,
            axistick_opts=opts.AxisTickOpts(is_show=True),
            splitline_opts=opts.SplitLineOpts(is_show=True),
        ),
        tooltip_opts=opts.TooltipOpts(is_show=True),
        legend_opts=opts.LegendOpts(pos_top="bottom")
    )
    .render_notebook()
)
```
