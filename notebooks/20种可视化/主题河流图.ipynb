```python

# 写一个函数！
def get_axis(var,year): 
    condi = f'{var}>0'
    df = sql(f'select {var} from {table}_{year} where {condi}') 
    y_data = []
    temp = []
    counts = df.value_counts()
    for i,n in counts.items():
        temp.append(f'{year}')
        temp.append(n)
        temp.append(info['range'][str(int(i[0]))])
        y_data.append(temp)
        temp = []
    return y_data
y_data = []
# y_data2 = get_axis(var,year2)
y_data.append(get_axis(var,year1))
y_data.append(get_axis(var,year2))
y_data.append(get_axis(var,year3))
y_data.append(get_axis(var,year4))
# y_data1.append(y_data2)
# y_data1
        
```

```python
data = []
for i in range(len(y_data)):
    for j in range(len(y_data[i])):
        data.append(y_data[i][j])
```

```python
# x = [f'{year1}',f'{year2}',f'{year3}',f'{year4}']
# x = sorted(x)
title = info['key']
x_data = []
for i,n in info['range'].items():
    if int(i)>0:
        x_data.append(n)  
```

```python
(
    ThemeRiver(init_opts=opts.InitOpts(width="100", height="800px"))
    .add(
        series_name=x_data,
        data=data,
        singleaxis_opts=opts.SingleAxisOpts(
            pos_top="50", pos_bottom="50", type_="time"
        ),
        label_opts=opts.LabelOpts(is_show=False)
    )
    .set_global_opts(
        tooltip_opts=opts.TooltipOpts(trigger="axis", axis_pointer_type="line"),
        title_opts=opts.TitleOpts(title=title, pos_left="center"),
        legend_opts=opts.LegendOpts(pos_top="bottom")
    )
    .render_notebook()
)
```
