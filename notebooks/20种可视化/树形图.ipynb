## 构造树

可以使用这里给出的自定义类来实现树的构造。

用`.add`方法添加一个节点，用`.echart`属性显示出树的`echart`格式

```python
class MyTree:
    def __init__(self, root_name: str='root', root_value: float=0):
        # note the parent of a vertex
        self.parents = [(0, None)]
        # dict: {id, name, value}
        self.vertices = [{'id': 0, 'name': root_name, 'value': root_value}]
    
    @property
    def root(self):
        return self.vertices[0]

    def add(self, parent: int, name: str, value: float=0):
        '''
        Add a vertex into the tree.
        parent: The id of the parent vertex.
        name: The name of the vertex.
        '''
        id_ = self.length
        self.parents.append((id_, parent))
        self.vertices.append({'id': id_, 'name': name, 'value': value})
        return self

    def __len__(self):
        return len(self.vertices)

    @property
    def length(self):
        return len(self.vertices)
    
    # A "branch" vertex is a vertex of degree at least 2
    # i.e., a branch vertex is a vertex which has children
    def is_branch(self, id_):
        '''
        Judge whether a vertex is branch
        '''
        return id_ in [parent for id__, parent in self.parents]
    
    def is_leaf(self, id_: int) -> bool:
        '''
        Judge whether a vertex is leaf
        '''
        return not self.is_branch(id_)

    def children(self, parent_id):
        return [id_ for id_, parent in self.parents if parent == parent_id]
    
    @property
    def echart(self):
        '''
        Change the tree into the format of echarts
        '''

        # Depth first search
        def dfs(vertex_id):
            echart_children = [dfs(child) for child in self.children(vertex_id)]
            vertex_info = self.vertices[vertex_id]
            echart = {
                'name': vertex_info['name'],
                'value': vertex_info['value'],
                'children': echart_children
            }
            return echart
        
        return [dfs(0)]
```

```python
mytree = MyTree('root', 100)
mytree.add(0, 'v1', 200).add(0, 'v2', 150).add(1, 'v3', 180).add(1, 'v4', 125)
tree = Tree().add('', mytree.echart)
tree.render_notebook()
```
