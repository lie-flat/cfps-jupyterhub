## 取出一列数据

```python
column = 'qe1' # 想要操作的列
```

```python
data = sql(f"SELECT {column} FROM adult_2010") # 使用 sql 函数可以执行查询，返回的是 pandas 的 DataFrame 格式
data
```

## 分组统计

```python
data_counts = data.value_counts()
data_counts
```
